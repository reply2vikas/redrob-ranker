## 2. Sandbox demo (small sample)
Graders run the ranker on a small sample (≤100 candidates). This cell makes a 100-row sample from the pool and ranks it end-to-end, proving reproducibility within budget.

```python
import glob, os, pandas as pd
hits = glob.glob('/content/**/submission.csv', recursive=True)
if not hits:
    raise SystemExit("submission.csv not found — re-run the ranking cell (Step 17) first; the runtime may have reset.")
src = hits[0]
print("Using:", src)
df = pd.read_csv(src)
df.to_csv('/content/TheVK7.csv', index=False)
df.to_excel('/content/TheVK7.xlsx', index=False)
print("Rows:", len(df))
from google.colab import files
files.download('/content/TheVK7.xlsx')
files.download('/content/TheVK7.csv')
```
